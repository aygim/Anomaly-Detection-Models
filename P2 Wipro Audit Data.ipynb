Isolation Forest (IF)
The Isolation Forest (iForest) algorithm is a tree-based ensemble method specifically designed for unsupervised anomaly detection. Unlike traditional techniques that model normal instances, iForest identifies anomalies by isolating them. The principle rests on the intuition that anomalies are â€œfew and differentâ€â€”they require fewer random partitions to be separated from the rest of the data. The model constructs multiple binary trees using random feature splits and measures the average path length required to isolate each observation; points with significantly shorter path lengths are deemed anomalous.
In the context of financial auditing, this approach is valuable because anomalies often manifest as irregular fluctuations in a small subset of metrics or time periods. iForest excels in identifying such deviations without requiring predefined thresholds or labelled training data.
The Isolation Forest model was implemented using Pythonâ€™s Scikit-learn library. Based on the domain understanding that less than 5% of financial records in a large, audited dataset are likely to contain irregularities, the contamination parameter was set to 0.05. The algorithm was trained on the standardized dataset, which included original metrics, ratios, and percentage-change features for the fiscal years 2021â€“2025.
 
The model produces binary outputs:
â€¢ 1= Normal observation
â€¢ -1 = Anomalous observation
The flagged fiscal year 2022 exhibits characteristics significantly different from its surrounding periods. When contextualized with Wiproâ€™s financial disclosures, this period corresponds to notable changes in Contingent Liabilities and Non-Current Borrowings, potentially linked to capital restructuring or asset revaluation events.
From an audit perspective, this result provides actionable insight. The Isolation Forest model does not identify the anomalyâ€™s cause but serves as an automated risk indicator. The auditor can then drill down into supporting schedules, working papers, or management notes for FY2022 to validate whether the anomaly stems from legitimate economic events or requires further audit evidence.
The use of Isolation Forest thus strengthens the audit process by enabling:
1. Population-level analysis instead of sampling-based checks.
2. Objective, data-driven flagging of unusual financial behaviour.
3. Efficient resource allocation, allowing auditors to focus on high-risk years or metrics first.

1. The Core Logic: "Finding the Normal"
The model's primary role is to find a boundaryâ€”a hyperplaneâ€”that encompasses the "normal" data points while separating them from the origin of the feature space.

Projection: It uses the Kernel Trick (specifically the RBF kernel in your code) to project your data into a high-dimensional space.

The Origin as Anomaly: In this high-dimensional space, it assumes the origin (zero-point) is the only "anomaly."

Maximum Margin: It calculates a boundary that is as far away from the origin as possible while still capturing the bulk of your data.

The One-Class Support Vector Machine (OC-SVM) is an unsupervised algorithm that learns the boundary of a single "normal" class in a high-dimensional feature space and identifies any data points lying outside this boundary as anomalies. Adapted from the classic Support Vector Machine framework, OC-SVM performs novelty detection. The model works by projecting data into a kernel-induced spaceâ€”commonly a Radial Basis Function (RBF) kernelâ€”where it constructs the smallest possible hypersphere enclosing most data points. Instances falling outside this learned boundary are treated as outliers.

From a financial auditing perspective, OC-SVM is valuable because it captures multivariate irregularitiesâ€”situations where a particular fiscal year may not appear anomalous in isolation but shows non-standard, potentially nonlinear relationships among multiple variables (e.g., simultaneous shifts in liabilities, reserves, and liquidity ratios). Unlike simple threshold-based tests, OC-SVM can detect complex, nonlinear deviations that auditors might overlook during traditional variance analysis.

The Local Outlier Factor (LOF) algorithm identifies anomalies by measuring the local deviation of a data pointâ€™s density compared to that of its neighbors.1 Unlike global models that assume a uniform distribution, LOF focuses on local density variationsâ€”assessing whether an observation lies in a region significantly less dense than its surroundings. This makes it particularly suited for detecting contextual anomalies, where an item is globally normal but unusual relative to similar peers.
In financial auditing, LOF is highly relevant because anomalies often arise locallyâ€”for example, a sudden shift in current liabilities or reserves that does not align with neighbouring yearsâ€™ proportional changes. By computing a "local density score," LOF enables auditors to pinpoint discrepancies that may not be captured through aggregate ratios or overall variance analysis.
The LOF algorithm was implemented using Scikit-learnâ€™s Local Outlier Factor class, which calculates the local reachability density (LRD) of each data point and compares it to the densities of its k-nearest neighbours. The degree of outlierness is expressed as the Local Outlier Factor score, where values significantly greater than 1 indicate local anomalies.
The model configuration used in this study was:
â€¢ neighbours = 2: The number of neighbouring observations considered for local density estimation. Given the short time series (five fiscal years), a smaller neighbourhood ensures sensitivity to short-term financial deviations.
â€¢ Contamination = 0.20: Specifies the expected fraction of anomalies, acknowledging that legitimate financial data may include exceptional fluctuations or events (e.g., acquisitions, debt restructuring).
â€¢ Input Data: The scaled analytical dataset containing both absolute financial metrics and year-over-year percentage changes.

The anomaly flagged in FY2022 corresponds with a sudden local deviation in the proportional relationship between Total Liabilities and Current Assets. Specifically, 2022 exhibited a comparatively higher liability ratio without a corresponding increase in assets, suggesting a temporary liquidity strain or reclassification of liabilities.
In practical audit terms, LOFâ€™s ability to detect such locally inconsistent years is invaluable. While Isolation Forest identifies broad, global anomalies and OC-SVM captures nonlinear structural deviations, LOF identifies peer-level irregularitiesâ€”those that stand out relative to adjacent fiscal periods. This aligns with how auditors often compare year-over-year balances and ratio movements: LOF automates that intuition, flagging years that behave abnormally compared to their immediate neighbours.

```python
# --- Synthetic anomaly injection ---

df_synth = X_scaled_df.copy()
# Synthetic Anomaly Injection - The goal here is to create a ground truth.
#Usually, unsupervised models (like Isolation Forest or One-Class SVM) operate "blindly"â€”they don't know which rows are actually bad. 
#By injecting synthetic shocks, you create a labeled validation set to measure your model's sensitivity.
shock_idx = np.random.choice(X_scaled_df.index, size=int(0.20*len(X_scaled_df)), replace=False)
#You are randomly selecting 20% of your rows to be transformed into anomalies. Using replace=False ensures you don't pick the same row twice.

df_synth.loc[shock_idx, 'Total Assets'] *= np.random.choice([1.7, 1.8, 1.2, 1.3], size=len(shock_idx))
#Injecting the "Shock"
#artificially inflating or deflating the Total Assets by 20â€“30%. 
#In financial modeling, a sudden 30% jump in assets without corresponding changes in other variables is a classic "red flag" behavior.

df_synth['synthetic_flag'] = 0
df_synth.loc[shock_idx, 'synthetic_flag'] = 1
#tells you exactly which rows you corrupted.

X_synth = df_synth[feature_cols]

df_synth['if_score_synth'] = -iso_model.decision_function(X_synth)
#You pass the corrupted data through your Isolation Forest (iso_model). You calculate the anomaly score for every row.

# Validation accuracy
synthetic_validation = df_synth.groupby('synthetic_flag')['if_score_synth'].mean()
synthetic_validation
#comparing the average anomaly score of normal rows vs. your injected rows.
```

Detection Strength: The model "noticed" your 30% asset shock immediately.

Threshold Clarity: You have a very clean "line in the sand." If you were to set a threshold at -0.01, you would catch almost all of your synthetic anomalies while ignoring almost all of your normal data.

```python
#performing Ensemble Anomaly Detection
#By averaging the scores of three different algorithms (Isolation Forest, Local Outlier Factor, and One-Class SVM) 
#reducing the bias of any single model
#creating a more robust "Consensus Score."
df_anomaly_results['Consensus_Score'] = (
    df_anomaly_results['IsolationForest_Score'] +
    df_anomaly_results['LOF_Score'] +
    df_anomaly_results['OneClass_SVM_Score']
) / 3

#weighted average method
#df_anomaly_results['Consensus_Score'] = (
    #2* df_anomaly_results['IsolationForest_Score'] +
    #1*df_anomaly_results['LOF_Score'] +
    #3*df_anomaly_results['OneClass_SVM_Score']
#) / 6

df_anomaly_results['Severity_Index'] = (
    (df_anomaly_results['Consensus_Score'] - df_anomaly_results['Consensus_Score'].min()) /
    (df_anomaly_results['Consensus_Score'].max() - df_anomaly_results['Consensus_Score'].min())
)

```

Adding a Rolling Standard Deviation is a smart move for time-series or ordered data. By calculating Score_Stability, we are essentially measuring the volatility of the anomaly signal.

A high stability score (high standard deviation) suggests that the anomaly is a sudden "spike" or "glitch," whereas a low stability score (low standard deviation) suggests a sustained, consistent shift in the data pattern.

```python
df_anomaly_results['Score_Stability'] = (
    df_anomaly_results['Consensus_Score']
    .rolling(window=3)
    .std()
)
```

Hyperparameter Sensitivity Analysis (Isolation Forest)
To assess the robustness of the Isolation Forest model to hyperparameter selection, a sensitivity analysis was conducted by varying the contamination parameter across three plausible levels (2%, 5%, and 10%). For each contamination level, the model was retrained on the standardized feature set, and the average anomaly score (negative decision function) was computed.
The contamination parameter controls the expected proportion of anomalies in the data. Evaluating model behaviour across multiple contamination settings ensures that the results are not unduly driven by a single, arbitrarily chosen threshold.
 
The average anomaly scores remain highly stable across all contamination levels, with only marginal variation observed as contamination increases. This stability indicates that the Isolation Forest modelâ€™s assessment of abnormality is robust to reasonable changes in hyperparameter specification.
From an audit analytics perspective, this result is important because it demonstrates that detected anomalies are not artifacts of parameter tuning but reflect structural characteristics of the financial data. The slight upward shift in average anomaly score at higher contamination levels is expected, as the model becomes more permissive in labelling anomalies; however, the magnitude of change is minimal, reinforcing confidence in the modelâ€™s reliability.

These discrepancies indicate that the anomalies identified by machine-learning approaches are not driven by extreme magnitude changes, but rather by subtle, multivariate, or structural irregularities that do not breach classical statistical thresholds.

This partial agreement demonstrates that while traditional statistical tests are effective at confirming the absence of extreme deviations, they may overlook complex anomaly patterns involving interactions across multiple financial variables.
